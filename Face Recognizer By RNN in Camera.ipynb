```python
# -*- coding: utf-8 -*-
# import usef libaries
import json
import cv2
from Facial_Train_By_RNN import Model
from LoadFaceData import resize_image, IMAGE_SIZE
```

```python
# Define a suitable class for face recognition.
class Face_recognition():
    
    def __init__(self):
        # Open the constrast table , output a name according to the number of name.
        with open('contrast_table', 'r') as f:
            self.contrast_table = json.loads(f.read())
            
        #Import the model and assign it to a class viariables.
        self.model = Model()
        
        # Load a trained model from a suitable path.
        self.model.load_model(file_path='./model')
        
        # Mark a face use a green frame.
        self.color = (0, 255, 0)

        # Catch a video stream from camera.
        self.cap = cv2.VideoCapture(0)

    def recongition(self):
        while True:
            # Read a frame picture
            ret, frame = self.cap.read()

            if ret is True:

                # Let the picture grey, which can reduce the complexity of process.
                frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            else:
                continue
                
            # Load the face claasifier.
            cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
            
            # Get the face use the detector.
            faceRects = cascade.detectMultiScale(frame_gray, scaleFactor=1.2, minNeighbors=3, minSize=(32, 32))
            if len(faceRects) > 0:
                for faceRect in faceRects:
                    x, y, w, h = faceRect

                    # Get the face picture
                    image = frame[y - 10: y + h + 10, x - 10: x + w + 10]
                    
                    # Input image into face_predicye function.
                    probability, name_number = self.model.face_predict(image)
                
                    # Use the return number to return a anme throught the constract table.
                    name = self.contrast_table[str(name_number)]

                    # print('name_number:', name_number)
                    # Specify the green frame, color and size.
                    cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), self.color, thickness=2)

                    # Output name in window
                    #cv2.putText(frame, name, (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)
                    if probability > 0.7:
                        cv2.putText(frame, name, (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)
                    else:
                        cv2.putText(frame, 'unknow', (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)

            cv2.imshow("face_recognition", frame)

            # Check if press any key after waiting 10s.
            k = cv2.waitKey(10)
            
            # Press Q to quit camera.
            if k & 0xFF == ord('q'):
                break

        # Release all windows and camera.
        self.cap.release()
        cv2.destroyAllWindows()


if __name__ == '__main__':
    fr = Face_recognition()
    fr.recongition()
```
