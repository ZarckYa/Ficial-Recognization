```python
import json
import os
import cv2
import numpy as np
```

```python
IMAGE_SIZE = 64

# Resize pecture according the specified size.
def resize_image(image, height=IMAGE_SIZE, width=IMAGE_SIZE):
    top, bottom, left, right = (0, 0, 0, 0)

    # Get the size of image.
    h, w, _ = image.shape  # (237, 237, 3)

    # Find the longest edge, if edge is not same.
    longest_edge = max(h, w)

    # Compute the short edge nned increase how many pixel dot.
    if h < longest_edge:
        dh = longest_edge - h
        top = dh // 2
        bottom = dh - top
    elif w < longest_edge:
        dw = longest_edge - w
        left = dw // 2
        right = dw - left
    else:
        pass

    # RGB
    BLACK = [0, 0, 0]

    # Edge fill 0 Adds a border to the image that is equal to the length and width of the image, 
    # cv2.BORDER_CONSTANT specifies the border colour specified by value
    constant = cv2.copyMakeBorder(image, top, bottom, left, right, cv2.BORDER_CONSTANT, value=BLACK)

    # Adjust and return.
    return cv2.resize(constant, (height, width))


def read_path(path_name):
    for dir_item in os.listdir(path_name):

        # Overlay from the initial path and merge into a recognisable operational path
        full_path = os.path.abspath(os.path.join(path_name, dir_item))

        # If it's a folder, continue the recursive call
        if os.path.isdir(full_path):
            read_path(full_path)
        else:
            if dir_item.endswith('.jpg'):
                image = cv2.imread(full_path)
                image = resize_image(image, IMAGE_SIZE, IMAGE_SIZE)

                # 放开这个代码，可以看到resize_image()函数的实际调用效果
                #cv2.imwrite('1.jpg', image)

                images.append(image)
                labels.append(path_name.split('\\')[-1])
    return images, labels


# Get data from specified path
def load_dataset(path_name):
    images = []
    labels = []
    labels1 = []
    images, labels = read_path(path_name)
    print('labels:', labels)


    # Convert all the input images into a 4-dimensional array with size (number of images * IMAGE_SIZE * IMAGE_SIZE * 3)
    # 600 images for me and my roommate, IMAGE_SIZE is 64, so for me the size is 1200 * 64 * 64 * 3
    # Images are 64 * 64 pixels, 3 colour values per pixel (RGB)
    images = np.array(images)
    print(images.shape)

    # label data.
    # get the number of different face.
    labels1 = list(set(labels))
    face_num = len(labels1)
    print(labels1)
    print('face_num:',face_num)
    
    # Store the test name into a dic called constrast_table. 
    num = [i for i in range(face_num)]
    contrast_table = dict(zip(num, labels1))
    with open('contrast_table', 'w') as f:
        f.write(json.dumps(contrast_table))
    print('contrast_table:', contrast_table)
    
    # Check the name and index.
    for index, name in contrast_table.items():
        for i in range(len(labels)):
            if labels[i] == name:
                labels[i] = index
    labels = np.array(labels)

    return images, labels, face_num


if __name__ == '__main__':
    images = []
    labels = []
    labels1 = []
    images, labels, _ = load_dataset("D:\Limerick course year 4\AI & FYP\Conda Jupyter notebook\My AI Folder\FYP\Facial Recognization By CNN\Facial data")
```
