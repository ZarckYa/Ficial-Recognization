```python
# -*- coding: utf-8 -*-
import os
import json
import cv2
import numpy as np
from FacialTrain import Model
from LoadFaceData import resize_image, IMAGE_SIZE
```

```python
#利用图片读取函数中返回的标签，将预测结果与标签对比，得到结果的正确和错误。将可以极大的增加数据结果的说服力。

# Define a suitable class for face recognition.
class Face_recognition():
    def __init__(self):
        
        # Open the constrast table , output a name according to the number of name.
        with open('contrast_table', 'r') as f:
            self.contrast_table = json.loads(f.read())
            
        #Import the model and assign it to a class viariables
        self.model = Model()
        
        # Load a trained model from a suitable path.
        self.model.load_model(file_path='./model')
        
        # Mark a face use a green frame.
        self.color = (0, 255, 0)

        # Local storage path for face recognition classifiers.
        self.cascade_path = "haarcascade_frontalface_default.xml"

    def read_Pictures(self, path_name):
        images = []
        labels = []
        for dir_item in os.listdir(path_name):
            # Iterarion from the initial path, until it is not folder, but file.
            full_path = os.path.abspath(os.path.join(path_name, dir_item))
            
            # If it is folder, continue open the folder.
            if os.path.isdir(full_path):
                self.read_Pictures(full_path)
            #If it is file and open the '.jpg' file.
            else:
                if dir_item.endswith('.jpg'):
                    image = cv2.imread(full_path)
                    image = resize_image(image, IMAGE_SIZE, IMAGE_SIZE)

                # 放开这个代码，可以看到resize_image()函数的实际调用效果
                #cv2.imwrite('1.jpg', image)

                    images.append(image)
                labels.append(path_name.split('\\')[-1])
        return images, labels

    def recongition(self, Path_name):
        images, labels = self.read_Pictures(Path_name)
        count_total = 0
        count_right = 0
        index = 0
        for image in images:
            # Let the picture grey, which can reduce the complexity of process.
            frame_gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
            print(np.array(frame_gray).shape)
            
            # Load the face claasifier
            cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
            
            # Get the face use the detector.
            faceRects = cascade.detectMultiScale(frame_gray, scaleFactor=1.2, minNeighbors=3, minSize=(32, 32))
            
            if len(faceRects) > 0:
                for faceRect in faceRects:
                    # Assign the face data to 4 variables.
                    x, y, w, h = faceRect
                    count_total += 1
                    
                    # Get the face picture
                    image_predict = image[y - 10: y + h + 10, x - 10: x + w + 10]
                    #print(image.shape)
                    
                    # Input image into face_predicye function.
                    probability, name_number = self.model.face_predict(image)
                    #name_number = self.model.face_predict(image)
                    #print(name_number)
                    
                    # Use the return number to return a anme throught the constract table.
                    name = self.contrast_table[str(name_number)]
                    print(name)
                
                    # Specify the green frame, color and size.
                    if name == labels[index]:
                        count_right += 1
                    index += 1 
                    # print('name_number:', name_number)
                    cv2.rectangle(image, (x - 10, y - 10), (x + w + 10, y + h + 10), self.color, thickness=2)

                    # Output name in window
                    cv2.putText(image, name, (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)
                    # if probability > 0.7:
                    #     cv2.putText(frame, name, (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)
                    # else:
                    #     cv2.putText(frame, 'unknow', (x + 30, y + 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)

            cv2.imshow("face_recognition", image)

            # Check if press any key after waiting 10s
            k = cv2.waitKey(10)

        # Release all windows and camera.
        #self.cap.release()
        cv2.destroyAllWindows()
        print("accuracy:", count_right/count_total)

if __name__ == '__main__':
    fr = Face_recognition()
    fr.recongition("D:\Limerick course year 4\AI & FYP\Conda Jupyter notebook\My AI Folder\FYP\Facial Recognization By CNN\Test face\Jufeng Yang")
```
