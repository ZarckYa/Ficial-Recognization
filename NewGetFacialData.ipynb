```python
import sys
import cv2
import tkinter as tk
from tkinter import messagebox
```

```python
Path_Location = ""

# Define class use to open windows, which use to comfirm if user need to collect new face.
class Comfirmation():
    def __init__(self):
        # Open a main window, title is comfirmation.size is 260*180.
        self.windows = tk.Tk()
        self.windows.title('Comfirmation')
        self.windows.geometry('260x180')
        
    # Setting for windows.add some lanel, button, tips.
    # If press the yes button, go to next window.
    # If press the no button, go to the exit tips.    
    def RequirementForNewFace(self):
        Info = tk.Label(self.windows, text = 'Do you want to collect new face data ?')
        Info.place(x = 10, y = 40)
        bt_yes = tk.Button(self.windows, text = 'Yes', command = self.Collection)
        bt_yes.place(x = 20, y = 140)
        bt_no = tk.Button(self.windows, text = 'No', command = self.Quit)
        bt_no.place(x = 160, y = 140)
        self.windows.mainloop()
    
    # Setting a exit tips.
    def Quit(self):
        messagebox.showinfo(title='Tips',message='Exit Successfully')
        self.windows.destroy()
    
    # Use to destiry the exist window.
    # Jump to the next function.
    def Collection(self):
        self.windows.destroy()
        MyCollectApp()
        
        
# A class to collect a new path.       
class MyCollectApp():
    def __init__(self):
        # Define a new main window.size is 280x120, title is Path Location.
        self.windows2 = tk.Tk()
        self.windows2.geometry('280x120')
        self.windows2.title('Path Location ')
        self.Path = []
        self.setupUI()
        #self.Path_Location = ""
    
    # Define a useful user interface.
    def setupUI(self):
        #define a Path label.
        label1 = tk.Label(self.windows2, text="Path：", height = 2, width = 10)
        label1.pack(side=tk.LEFT)
        
        # Define the varialble: path as a string type.
        self.Path = tk.StringVar()
        
         # Entry a path. used by the follow code.
        tk.Entry(self.windows2, textvariable = self.Path).place(x = 55, y = 50)
        Comfirm_Button = tk.Button(self.windows2, text="Comfirm", command=self.on_click).place(x = 100, y = 90)
        
    def on_click(self):
        # Define a global Variable, which used to pass the path into follow function.
        global Path_Location
        Path_Location = self.Path.get().lstrip()
        
        # Check if the input is empty. if it is empty, Output a warning message.
        if len(Path_Location) == 0:
            messagebox.showwarning(title='System Warning',message='Please input a valid path!')
            return False
        
        # Quit windows and destroy  windows.
        self.windows2.quit()
        self.windows2.destroy()
        print("Path：%s" % (Path_Location))
        return Path_Location

def CatchPICFromVideo(path_name, window_name="GET_FACE", camera_idx=0, catch_pic_num=1500):
    cv2.namedWindow(window_name)

    # Open a camera, define a funtion into a variables cap.
    cap = cv2.VideoCapture(camera_idx)

    # Use a face detector classefier to capyure a face.
    classfier = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")

    # Produce a green color frame.
    color = (0, 255, 0)

    num = 0
    while cap.isOpened():
        ok, frame = cap.read()
        #print(type(frame))
        if not ok:
            break
        
        # Convert present data into grey picture data
        grey = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

        # Face detection, 1.2 and 2 are the image scaling and the number of valid points to be detected respectively
        faceRects = classfier.detectMultiScale(grey, scaleFactor=1.2, minNeighbors=3, minSize=(32, 32))
        
        # If the faceRects is bigger than 0, It means face is detected successfully
        if len(faceRects) > 0: 
            
            # Go though each face picture.
            for faceRect in faceRects:
                x, y, w, h = faceRect

                # Save present face data into a path.
                img_name = '%s/%d.jpg ' % (path_name, num)
                image = frame[y - 10: y + h + 10, x - 10: x + w + 10]
                cv2.imwrite(img_name, image)

                 # Counter: record how many face picture catch and saved.
                # If num bigger than the MAX capture number, Exit loop and stop catch and save.
                num += 1
                if num > (catch_pic_num):
                    break

                # Draw a green rectangle frame,
                cv2.rectangle(frame, (x - 10, y - 10), (x + w + 10, y + h + 10), color, 2)

                # Show how many pictures has been stored.
                font = cv2.FONT_HERSHEY_SIMPLEX
                cv2.putText(frame, 'num:%d' % (num), (x + 30, y + 30), font, 1, (255, 0, 255), 4)

                # Over the maximum picture number, quit the loop 
        if num > (catch_pic_num): break

        # show picture.
        cv2.imshow(window_name, frame)
        c = cv2.waitKey(10)
        if c & 0xFF == ord('q'):
            break

    # Release camera, and destory windows.
    cap.release()
    cv2.destroyAllWindows()

if __name__ == '__main__':
    Cr = Comfirmation()
    Cr.RequirementForNewFace()
    if Path_Location != "":
        CatchPICFromVideo(Path_Location)
    print(Path_Location)
```
